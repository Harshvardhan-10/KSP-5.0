```python
def get_data(VLASS_fits, WISE_fits, PanSTARRS_fits):
    VLASS_data = VLASS_fits[0].data[0,0,:,:]
    WISE_data = WISE_fits[0].data
    PanSTARRS_data = PanSTARRS_fits[0].data
    return VLASS_data, WISE_data, PanSTARRS_data
```

```python
def clipped_plotter(VLASS_data, WISE_data, PanSTARRS_data, title):

    mean_VLASS, median_VLASS, std_VLASS = sigma_clipped_stats(hydraA_VLASS_data)
    mean_WISE, median_WISE, std_WISE = sigma_clipped_stats(hydraA_WISE_data)
    mean_PanSTARRS, median_PanSTARRS, std_PanSTARRS = sigma_clipped_stats(hydraA_PanSTARRS_data)
    
    fig, ax = plt.subplots(1,3,figsize=(15,5))
    color = 'magma'

    ax[0].imshow(VLASS_data, vmin = median_VLASS - 10*std_VLASS, vmax = median_VLASS + 10*std_VLASS, cmap=color)
    ax[0].title.set_text('VLASS')
    # ax[0].colorbar()

    ax[1].imshow(WISE_data, vmin = median_WISE - 10*std_WISE, vmax = median_WISE + 10*std_WISE, cmap=color)
    ax[1].title.set_text('WISE')
    # ax[1].colorbar()

    ax[2].imshow(PanSTARRS_data, vmin = median_PanSTARRS - 10*std_PanSTARRS, vmax = median_PanSTARRS + 10*std_PanSTARRS, cmap=color)
    ax[2].title.set_text('PanSTARRS')
    # ax[2].colorbar()

    plt.suptitle(title)
    plt.tight_layout()
    plt.show()
```

```python
def plotter(VLASS_data, WISE_data, PanSTARRS_data, title):

    fig, ax = plt.subplots(1,3,figsize=(15,5))
    color = 'magma'

    ax[0].imshow(VLASS_data, cmap=color)
    ax[0].title.set_text('VLASS')
    # ax[0].colorbar()

    ax[1].imshow(WISE_data, cmap=color)
    ax[1].title.set_text('WISE')
    # ax[1].colorbar()

    ax[2].imshow(PanSTARRS_data, cmap=color)
    ax[2].title.set_text('PanSTARRS')
    # ax[2].colorbar()

    plt.suptitle(title)
    plt.tight_layout()
    plt.show()
```

```python
plotter(hydraA_VLASS_data, hydraA_WISE_data, hydraA_PanSTARRS_data, 'HydraA')
clipped_plotter(hydraA_VLASS_data, hydraA_WISE_data, hydraA_PanSTARRS_data, 'HydraA')
```

Assuming you are an observer located at the GMRT site in Narayangaon, Pune (latitude of about +19 deg N).

### 1. What is the altitude and azimuth of North Celestial Pole?
Altitude = $+19 \degree$ \\
Azimuth = $0 \degree$

### 2. What is the altitude at transit of an object on the celestial equator (Dec=0 deg)?
Altitude = 90 - (19 - 0) = $71\degree$

### 3. What is the azimuth of a star on the celestial equator at the moment that it is setting?
Star sets on the west at the celestial equator. Hence Azimuth = 270 $\degree$

### 4. Assuming GMRT can observe everything above the horizon, which of the following can never be observed with GMRT - M45, M83, Omega Centauri, Cassiopeia A, Centaurus A, Large Magellanic Cloud, Andromeda Galaxy, Carina Nebula?
GMRT is at Latitude $+19\degree$. To be able to observe something its Dec needs to be between 90 - Lat and -90 + Lat degree.

So the GMRT can observe everything in between the Dec of $+90\degree$ and $-71\degree$.

Dec data for: \
M45: $+2406'$ \
M83: $29 51' 56.74"$ \
Omega Centauri: $-47\degree 29'$\
Cassiopeia A: $+58 48.9'$\
Centaurus A: $-43 1' 9"$\
Large Magellanic Cloud: $-69 45' 22"$ \
Andromeda Galaxy: $+41 16' 9"$ \
Carina Nebula:  $59 52' 04"$

So the GMRT can observe all the celestial bodies, but practically it will be difficult for it to observe the Large Magellanic Cloud as it will be very low in the sky.

### 5. GW170817 (named by discovery date - 17 August 2017) was the first ever detection in electromagnetic waves of a compact binary merger event of two neutron stars following its detection in gravitational waves with LIGO (More info : https://www.ligo.caltech.edu/page/press-release-gw170817). You are tasked with obtaining radio data of GW170817 (RA 13h 09m 48.08s and Dec -23d 22m 53.3s). 


### What is the expected altitude of GW170817 when it is transiting?

$Altitude = 90 - |Lat-Dec|$
$\\ Altitude = 90 - |19 - (-23\degree 22' 53.3'')| \\ Altitude = 90 - |19 - (-23.3815)| = 90 - |19 + 23.3815| = 90 - 42.3815 = 47.6185 \degree$


### 6. Bonus : Radio telescopes are capable of observing even during the day as long as the source not close to the Sun in the sky. Check the position of the Sun during GW170817 - can you comment on its visibility during that period? Can you think of some specific time of the year when it would not be possible to observe GW170817?

Coordinates of GW170817: RA 13h 09m 48.08s and Dec -23d 22m 53.3s \\
Coordinates of The Sun during GW170817: Right Asc: 09h 46m 38.6s Decl: +13 24' 02.5" (J2000)

The Sun was not close to GW170817 on 17/8/17 and hence Radio data of it could be observed on that day.

The Sun migrates between the Tropic of Cancer($23.5\degree$) and Tropic of Capricorn($-23.5\degree$) every year. So around the Winter Solstice (23rd Dec) the Sun would be on the same Dec as GW170817 making it difficult to observe.

```python
from astropy.io import ascii
gw170817 = ascii.read('gw170817_afterglow_data_full.txt')
```

```python
gw170817
```

```python
row = []
i = 0
for flux in gw170817['FluxD']:
    if '<' not in flux:
        pass
    else:
        row.append(i)
    i = i+1
```

```python
gw170817.remove_rows(row)
```

```python
gw170817
```

```python
gw170817_VLA = gw170817[gw170817['Telescope'] == 'VLA']
gw170817_VLA_3GHz = gw170817_VLA[gw170817_VLA['Freq'] == 3000000000]
```

```python
gw170817_VLA_3GHz
```

```python
x = np.array(gw170817_VLA_3GHz['T'])
y = np.array(gw170817_VLA_3GHz['FluxD'], dtype=float)

yerr = np.array(gw170817_VLA_3GHz['FluxDErr'])
```

```python
plt.figure(figsize = (10,6))
plt.title('Light Curve')
plt.xlabel('Time After Merger (Days)')
plt.ylabel('Flux Density (Jy)')
plt.errorbar(x,y,yerr, fmt = 'o')
```

```python
gw170817_Chandra = gw170817[gw170817['Telescope'] == 'Chandra']
gw170817_Chandra['FluxD'] = gw170817_Chandra['FluxD'].astype(float)
# Rescaling the values to 3GHz using the proportionality relation F_v proportional to v^-0.584
gw170817_Chandra['FluxD Scaled'] = gw170817_Chandra['FluxD'] * np.power((3*(10**9))/(2.41*(10**17)), -0.584)
gw170817_Chandra['FluxDErr Scaled'] = gw170817_Chandra['FluxDErr'] * np.power((3*(10**9))/(2.41*(10**17)), -0.584)
```

```python
gw170817_Chandra
```

```python
x_Chandra = np.array(gw170817_Chandra['T'])
y_Chandra = np.array(gw170817_Chandra['FluxD Scaled'])
y_Chandra_err = np.array(gw170817_Chandra['FluxDErr Scaled'])
```

```python
plt.figure(figsize = (10,6))
plt.title('Light Curve')
plt.xlabel('Time After Merger (Days)')
plt.ylabel('Flux Density (Jy)')
plt.errorbar(x_Chandra,y_Chandra, y_Chandra_err, fmt = 'o', label = 'Chandra') 
plt.errorbar(x,y,yerr, fmt = 'o', label = 'VLA')
plt.legend()
plt.show()
```
